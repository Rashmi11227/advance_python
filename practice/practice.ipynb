```python
class replace_missing_con:
    def __init__(self,data):
        self.data=data
    def replace(self):
        print(f"missing values before replace {(self.data).isna().sum()}")    
        med=(self.data).median()
        (self.data).fillna(med,inplace=True)
        print(f"missing values after replaced {(self.data).isna().sum()}")
        print(f"missing values replaced with {med}")
    
```

```python
class BaseSalary:
    def __init__(self,salary):
        self.salary=salary
    def get_base_salary(self):
        return self.salary    
```

```python
class Bonus:
    def __init__(self,bonus_per):
        self.bonus_per=bonus_per

    def calculate_bonus(self,base_salary):
        return (self.bonus_per/100)*base_salary

        
```

```python
class Tax:
    def __init__(self,tax_per):
        self.tax_per=tax_per
    def calculate_tax(self,gross_salary):
        return (self.tax_per/100)*gross_salary    
```

```python
class TotalSalary(BaseSalary,Bonus,Tax):
    def __init__(self,salary,bonus_per,tax_per):
        BaseSalary.__init__(self,salary)
        Tax.__init__(self,tax_per)
        Bonus.__init__(self,tax_per)
    def calculate_total_salary(self):
          base_sal=self.get_base_salary()
          bonus=self.calculate_bonus(base_sal)
          gross_sal=base_sal+bonus
          tax=self.calculate_tax(gross_sal)
          net_sal=gross_sal-tax
          return {
               "base_salary":base_sal,
               "Bonus":bonus,
               "Gross_Salary":gross_sal,
               "Tax":tax,
               "Net_Salary":net_sal

          }
```

```python
class Employee:
    def __init__(self,name,emp_id):
        self.name=name
        self.emp_id=emp_id
    def info(self):
        return f"Name={self.name} and Employee_Id={self.emp_id}"
```

```python
class Manager(Employee):
    def __init__(self, name, emp_id,department):
        super().__init__(name, emp_id)
        self.department=department
    def manager_info(self):
        return f"{self.info()}, Depatment={self.department}"  
```

```python
class Project_Manager(Manager):
    def __init__(self, name, emp_id, department,project):
        super().__init__(name, emp_id, department)
        self.project=project
    def get_project_info(self):
        return f"{self.manager_info()} Project={self.project}"    
```
