## multilevel inheritance((parent,child,grandchild)


```python
class employee:
    def __init__(self,name,emp_id):
        self.name=name
        self.emp_id=emp_id
    def get_employee_info(self):
        return f"name={self.name},ID={self.emp_id}"
```

```python
class manager(employee):
    def __init__(self, name, emp_id,department):
        super().__init__(name, emp_id)
        self.department=department
    def get_manager_info(self):
        return f"{self.get_employee_info()},Department={self.department}"    
```

```python
class project_manager(manager):
    def __init__(self, name, emp_id, department,project):
        super().__init__(name, emp_id, department)
        self.project=project
    def get_project_info(self):
        return f"{self.get_manager_info()},Project={self.project}"    
```

```python
emp4.emp_id
```

```python
emp5=manager("suman",234,"IT")
```

```python
class parent:
    def __init__(self,name,surname):
        self.name=name
        self.surname=surname
    def introduce(self):
        print(f"Hi myself {self.name} and my surname is {self.surname}")    
```

```python
class child(parent):
    def __init__(self, name, surname,address):
        super().__init__(name, surname)
        self.address=address
    def child_introduce(self):
        super().introduce()
        print(f"My address is {self.address}")
      
```

```python
a1=parent("suman","patil")
```

```python
a1.name
```

```python
a1.surname
```

```python
a1.introduce()
```

```python
a2=child("raman","patil","wakad")
```

```python
a2.name
```

```python
a2.introduce()
```

```python
a2.child_introduce()
```

```python
class BaseSalary:
    def __init__(self,salary:float):
        self.salary=salary
    def get_base_salary(self):
        return self.salary    
```

```python
class Bonus:
    def __init__(self,Bonus_Per:float):
        self.Bonus_Per=Bonus_Per
    def calculate_bonus(self,base_salary:float)->float:
        return (self.Bonus_Per/100)*base_salary  
```

```python
class Tax:
    def __init__(self,Tax_Per:float):
        self.Tax_Per=Tax_Per
    def calculate_tax(self,gross_Sal:float)->float:
        return (self.Tax_Per/100)*gross_Sal    
```

```python
class TotalSalary(BaseSalary,Bonus,Tax):
    def __init__(self,salary:float,Bonus_Per:float,Tax_Per:float):
        BaseSalary.__init__(self,salary)
        Bonus.__init__(self,Bonus_Per)
        Tax.__init__(self,Tax_Per)
    def calculate_net_Salary(self)->dict[str,float]:
        base_sal=self.get_base_salary()
        bonus=self.calculate_bonus(base_sal)
        gross_sal=base_sal+bonus
        tax=self.calculate_tax(gross_sal)
        net_salary=gross_sal-tax
        return {
            "basesalary":base_sal,
            "bonus":bonus,
            "gross_salary":gross_sal,
            "tax":tax,
            "net_Salary":net_salary
        }

                            
```

```python
s2=TotalSalary(salary=50000,Bonus_Per=10,Tax_Per=12)
```

```python
s2.salary
```

```python
s2.get_base_salary()
```

```python
s2.calculate_net_Salary()
```
